### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
#Figure: objekt där diagramet ritas upp på. Håller koll på allt som diagrammet består av. Linjer, prickar, staplar, titlar, labels. 
#Axes: Diagrammet ritas på ett objekt som kallas Axes (själva plottområdet). En Figure kan bestå av en eller flera Axes. Det primära gränssnittet för att få ett diagram. Data, skala på axlar och begränsningar, lägga på etiketter.
#Axis: Linjer (x-axeln, y-axeln, z-axeln) som definierar koordinatsystemet, inklusive skalor och tick-märken, inom det området.
#Artist: Allt synligt i en Figure är en Artist. Det kan vara en linje som representerar data, en text etikett, grid i ett diagram, eventuella markeringar etc.  
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#OO-style

x = np.linspace(-4, 4) #-4<= x <=4 
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y, '*-') #mark on each value
ax.set_title('y = x^2') # title head
ax.set_xlabel('x') #label for x-axis
ax.set_ylabel('y') #label for y-axis
```

```python
#pyplot approach

x = np.linspace(-4, 4) #-4<= x <=4 
y = x**2

plt.plot(y)

plt.title('Implicit plot, y = x^2')
plt.xlabel("x")
plt.ylabel("y")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, axs = plt.subplots(2)
axs[0].scatter(random_data_x, random_data_y)
axs[1].bar(names, values)
```
